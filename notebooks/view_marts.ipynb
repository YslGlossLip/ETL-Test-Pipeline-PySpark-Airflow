```python
from pathlib import Path
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

BASE_DIR = Path.cwd()

if BASE_DIR.name == "notebooks":
    BASE_DIR = BASE_DIR.parent

print("Project root:", BASE_DIR)
print("Data folder exists:", (BASE_DIR / "data").exists())

spark = (
    SparkSession.builder
    .appName("view_data")
    .getOrCreate()
)

spark.sparkContext.setLogLevel("ERROR")
```

```python

by_hardware_path = BASE_DIR / "data" / "gold" / "mart_by_hardware"
by_mm_dma_path = BASE_DIR / "data" / "gold" / "mart_by_mm_dma"
by_site_id_path = BASE_DIR / "data" / "gold" / "mart_by_site_id"

viewing = by_hardware_path

print("Path:", viewing)
print("Exists:", viewing.exists())

viewing_mart = spark.read.parquet(str(viewing))

# print(str(viewing))
print("Mart Rows:", viewing_mart.count())
viewing_mart.printSchema()

viewing_mart.orderBy("reg_date", "hardware").limit(100).toPandas()
# viewing_mart.show(20, truncate= False)

# spark.stop()
```
